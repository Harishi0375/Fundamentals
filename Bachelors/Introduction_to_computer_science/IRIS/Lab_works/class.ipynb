```python
#importing the libararies
import numpy as np
import matplotlib.pyplot as plt  
```

```python
#create numpy arrays
x = np.array([1,2,3,4])
y = np.array([4,5,6,7])

#plot the graph
plt.plot(x,y, color="red", linestyle="dashed")

#add the labels, title and grid
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Title")
plt.grid()

plt.show()
```

```python
#creating multiple axes subplots with 2 rows and 2 columns
fig, ax = plt.subplots(2,2)

#creating multiple arrays
x = np.array([1,2,3,4])
y = np.array([4,5,6,7])
a = np.array([6,3,8,2])
b = np.array([4,7,3,9])

#plotting all the graphs
ax[0][0].plot(x,y)
ax[0][1].plot(b,a)
ax[1][0].plot(x,a)
ax[1][1].plot(y,b)

plt.show()
```

```python
#plotting bar graph
x = np.array(["a","b","c","d"])
y = np.array([4,5,6,7])

plt.bar(x,y, width=0.5)

plt.show()
```

```python
#plotting scatter plot
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

```python
#creating numbers for 0 to 100 with step of 2
x = np.linspace(0, 2 ,100)

fig, ax = plt.subplots()

#plotting all three graphs
ax.plot(x, x, label="Linear")
ax.plot(x, x**2, label="quad")
ax.plot(x, x**3, label="cubic")

#adding the legend
ax.legend()

plt.show()
```
