# Plotting

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the CSV file into a DataFrame
raw_bottom_data = pd.read_csv("./data/bottom_data.csv", header=1, usecols=[1])
raw_closed_data = pd.read_csv("./data/closed_data.csv", header=1, usecols=[1])
raw_matrix_data = pd.read_csv("./data/matrix_data.csv", header=1, usecols=[1])
raw_naiv_data = pd.read_csv("./data/naiv_data.csv", header=1, usecols=[1])
```

```python
bottom_x = np.arange(0, len(raw_bottom_data), 1)
bottom_y = raw_bottom_data.values.flatten()

closed_x = np.arange(0, len(raw_closed_data), 1)
closed_y = raw_closed_data.values.flatten()

matrix_x = np.arange(0, len(raw_matrix_data), 1)
matrix_y = raw_matrix_data.values.flatten()

naiv_x = np.arange(0, len(raw_naiv_data), 1)
naiv_y = raw_naiv_data.values.flatten()
```

```python
plt.tight_layout()
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(bottom_x, bottom_y, label="Bottom-up", color="red")
axs[0, 0].set_title("Bottom-up")
axs[0, 0].set_yscale("log")

axs[0, 1].plot(closed_x, closed_y, label="Closed-form", color="green")
axs[0, 1].set_title("Closed-form")
axs[0, 1].set_yscale("log")

axs[1, 0].plot(matrix_x, matrix_y, label="Matrix", color="orange")
axs[1, 0].set_title("Matrix")
axs[1, 0].set_yscale("log")

axs[1, 1].plot(naiv_x, naiv_y, label="Naive", color="purple")
axs[1, 1].set_title("Naive")
axs[1, 1].set_yscale("log")

for ax in axs.flat:
    ax.set(xlabel="fib nth term", ylabel="fib value")
    ax.legend()

plt.tight_layout()
```
