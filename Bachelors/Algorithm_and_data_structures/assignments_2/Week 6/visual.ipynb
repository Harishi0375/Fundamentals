First I need to load the data and visualize it on the same plot

```python
import pandas as pd
import numpy as np

# Import the data from the CSV file
bottom = pd.read_csv("./Data/bottom_up_heap_sort.csv", header=1, usecols=[1])
regular = pd.read_csv("./Data/Regular_heap_sort.csv", header=1, usecols=[1])

# Extract the second column
bottom_y = bottom.values.flatten()
regular_y = regular.values.flatten()

x = np.arange(0, len(bottom), 1)
```

```python
import matplotlib.pyplot as plt

# Plot the dataframes
plt.plot(x, bottom_y, label='Bottom-up heap sort')
plt.plot(x, regular_y, label='Regular heap sort')

# Set the plot title and labels
plt.title('Comparison of bottom and regular data')
plt.xlabel('size of array')
plt.ylabel('time taken (ms)')

# Set the y-axis limits


# Display the legend
plt.legend()

# Show the plot
plt.savefig('heap_sort-comparison.png')

```

Fromm the data, the difference is not as vast, but what is obvious is that the bottom up algorithm, is slightly more efficent than the regular heap sort taking it around 200-300 more elements to reach the same exection time as the regular Heap Sort
