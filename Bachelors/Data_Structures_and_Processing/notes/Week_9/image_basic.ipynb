```python
import imageio.v3 as iio
import ipympl
import matplotlib.pyplot as plt
import numpy as np
import skimage as ski
```

#### Pixels, colours, and arrays


```python
image = np.array([
[0, 0, 0],
[0, 1, 0], #where 1 = different colour
[0, 0, 0],
[0, 1, 0], #where 1 = different colour k
[0, 0, 0]
])

plt.imshow(image)
plt.show()
```

##### To change colour


```python
plt.imshow(image, cmap = plt.cm.gray)  #change colour
plt.show()
```

Images represented in 2-dimensional are called monochromatic images. Usually, true colour images are represented by 3-dimensional arrays, where each colour channel is represented by a 2D array. The most common standard for colour images is the additive RGB where each pixel is represented by three values: Red, Green, and Blue. Values range from 0 to 255. Try the following:

```python
rgen = np.random.RandomState(2024) 
# 2024 values that repeat the same everytime (no new vales each time)
img = rgen.randint(0,255,size=(4,4,3))
plt.imshow(img)
plt.show()
```

For red channel:

```python
red_channel = img * [1, 0, 0]
plt.imshow(red_channel)
plt.show()
```

For green channel:

```python
green_channel = img * [0, 1, 0]
plt.imshow(green_channel)
plt.show()
```

For blue channel:


```python
blue_channel = img * [0, 0, 1]
plt.imshow(blue_channel)
plt.show()
```
